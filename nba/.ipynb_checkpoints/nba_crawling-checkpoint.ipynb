```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from pandas import *
import time
```

```python
teams = ['ATL', 'BOS', 'BRK', 'CHO', 'CHI', 'CLE', 'DAL', 'DEN', 'DET', 'GSW', 'HOU', 'IND', 'LAC', 'LAL',
                 'MEM', 'MIA', 'MIL', 'MIN', 'NOP', 'NYK', 'OKC', 'ORL', 'PHI', 'PHO', 'POR', 'SAC', 'SAS', 'TOR',
                 'UTA', 'WAS']
team_player = {}
team_player_url = {}
dr = webdriver.Chrome('/Users/joono/chromedriver')
for team in teams:
    url = 'https://www.basketball-reference.com/teams/'+team+'/2019.html'
    dr.get(url)
    players = []
    players_url = []
    players_el = dr.find_elements_by_css_selector('td[data-stat = "player"]')
    for pl_el in players_el:
        players.append(pl_el.text)
        players_url.append(pl_el.find_element_by_tag_name('a').get_attribute('href'))
    if players == []:
        print(team,'잘못됨')
    team_player.update({team:list(unique(players))})
    team_player_url.update({team:list(unique(players_url))})
```

```python
G = []
Date = []
Age = []
Tm =[]
at=[]
Opp=[]
wl=[]
GS=[]
MP=[]
FG=[]
FGA=[]
FGper=[]
three_P=[]
three_PA=[]
three_Pper=[]
FT=[]
FTA=[]
FTper=[]
ORB=[]
DRB=[]
TRB=[]
AST=[]
STL=[]
BLK=[]
TOV=[]
PF=[]
PTS=[]
GmSc=[]
plus=[]
stats = [G,
Date,
Age,
Tm,
at,
Opp,
wl,
GS,
MP,
FG,
FGA,
FGper,
three_P,
three_PA,
three_Pper,
FT,
FTA,
FTper,
ORB,
DRB,
TRB,
AST,
STL,
BLK,
TOV,
PF,
PTS,
GmSc,
plus]
team_col = []
name_col = []
dr = webdriver.Chrome('/Users/joono/chromedriver')
nowtime time.time()
for team in teams[:1]:
    urls = [x[:-5]+'/gamelog/2018' for x in team_player_url[team]]
    for idx,url in enumerate(urls[:1]):
        dr.get(url)
        table = dr.find_element_by_css_selector('table[class = "row_summable sortable stats_table now_sortable"]')
        tbody = table.find_element_by_css_selector('tbody')
        trs = tbody.find_elements_by_css_selector('tr')
        for tr in trs:
            tds = tr.find_elements_by_css_selector('td')
            if len(tds) != 0:
                if tds[0].text != '':
                    for i in range(0,len(stats)):
                        try:
                            if tds[i].text != '':
                                stats[i].append(tds[i].text)
                            else:
                                stats[i].append('none')
                        except IndexError:
                            stats[i].append('none')

                else:
                    for i in range(0,len(stats[:8])):
                        try:
                            if tds[i].text != '':
                                stats[i].append(tds[i].text)
                            else:
                                stats[i].append('no_play')
                        except IndexError:
                            stats[i].append('no_play')

                    for j in range(8,len(stats)):
                        stats[j].append('no_play')
                team_col.append(team)
                name_col.append(team_player[team][idx])
            else:
                continue
dt = DataFrame({'G' : G,
            'Date' : Date,
            'Age' : Age,
            'Tm' : Tm,
            'at' : at,
            'Opp' : Opp,
            'wl' : wl,
            'GS' : GS,
            'MP' : MP,
            'FG' : FG,
            'FGA' : FGA,
            'FGper' : FGper,
            'three_P' : three_P,
            'three_PA' : three_PA,
            'three_Pper' : three_Pper,
            'FT' : FT,
            'FTA' : FTA,
            'FTper' : FTper,
            'ORB' : ORB,
            'DRB' : DRB,
            'TRB' : TRB,
            'AST' : AST,
            'STL' : STL,
            'BLK' : BLK,
            'TOV' : TOV,
            'PF' : PF,
            'PTS' : PTS,
            'GmSc' : GmSc,
            'plus' : plus,
            'team' : team_col,
           'name' : name_col})
endtime = time.time() - nowtime
print(endtime)
print('done')
```
